```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# For Your Eyes Only

frizzy = cv2.imread('images/frizzy.png')
plt.imshow(frizzy)
```

```python
froomer = cv2.imread('images/froomer.png')
plt.imshow(froomer)
```

```python
# TODO: Find edges in frizzy and froomer images

images = {'frizzy':frizzy, 'froomer':froomer}
edges = {}

thresh_hi = 110
thresh_lo = 60
thresh = (thresh_hi, thresh_lo)

for key in images.keys():
    edge = cv2.Canny(images[key], *thresh)
    edges[key] = edge
    plt.figure()
    plt.imshow(edge, cmap='binary_r')
    
```

```python
# TODO: Display common edge pixels
def edge_intersect(img1, img2):
    return edges[img1] & edges[img2]
```

```python
plt.imshow(edge_intersect('frizzy', 'froomer'), cmap='binary_r')
```
