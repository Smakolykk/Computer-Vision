```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('images/fall-leaves.png')
img = img[:, :, ::-1]
```

```python
plt.imshow(img)
```

```python
size = 21
sigma = 9

kernel = cv2.getGaussianKernel(size, sigma)
kernel2D = kernel * kernel.T
```

```python
plt.matshow(kernel2D)
plt.colorbar()
```

```python
plt.imshow(cv2.filter2D(img, -1, kernel2D, borderType=cv2.BORDER_CONSTANT))
```

```python
plt.imshow(cv2.filter2D(img, -1, kernel2D, borderType=cv2.BORDER_REPLICATE))
```

```python
plt.imshow(cv2.filter2D(img, -1, kernel2D, borderType=cv2.BORDER_REFLECT))
```

```python
plt.imshow(cv2.filter2D(img, -1, kernel2D, borderType=cv2.BORDER_REFLECT_101))
```

```python
plt.imshow(cv2.filter2D(img, -1, kernel2D, borderType=cv2.BORDER_DEFAULT))
```

```python
plt.imshow(cv2.filter2D(img, -1, kernel2D, borderType=cv2.BORDER_ISOLATED))
```
