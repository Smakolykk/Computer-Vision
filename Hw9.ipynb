```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Gaussian Kernel params
size  = 21
sigma = 9


def downsample(img):
    # TODO: img_d = ? (pick alternate rows, cols: 1, 3, 5, ...)
    return img[ np.arange(1, img.shape[0], 2) ]


def blur_downsample(img):
    # TODO img_bd = ? (blur by 5x5 gaussian, then downsample)
    kernel = cv2.getGaussianKernel(size, sigma)
    kernel2D = kernel*kernel.T
    filtered_img = cv2.filter2D(img, -1, kernel2D)
    return downsample(filtered_img)
```

```python
img = cv2.imread('images/frizzy.png') [:, :, ::-1]
plt.imshow(img)
print(img.shape)
```

```python
# downsample image
img_d = downsample(img)
img_d = downsample(img_d)
img_d = downsample(img_d)
print(img_d.shape)
```

```python
# blur and downsample
img_bd = blur_downsample(img)
img_bd = blur_downsample(img_bd)
img_bd = blur_downsample(img_bd)
print(img_bd.shape)
```

```python
plt.figure()
plt.imshow(cv2.resize(img_d, (img.shape[1], img.shape[0])))

plt.figure()
plt.imshow(cv2.resize(img_bd, (img.shape[1], img.shape[0])))
```
