```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def compute_integral(img):
    # TODO: Compute I such that I(y,x) = sum of img(1,1) to img(y,x)
    integral = np.zeros_like(img)
    for i,j in zip(range(img.shape[0]), range(img.shape[1])):
        integral[i][j] = np.sum(img[:i, :j])
    return integral
```

```python
size = (8,8)

img = cv2.imread('images/dolphin.png', 0)

plt.figure(figsize=(size))
plt.imshow(img, cmap='binary_r')
print(img.shape)
```

```python
# compute integral
img = np.float64(img)
I = compute_integral(img)

plt.figure(figsize=(size))
plt.imshow((I / I.max()), cmap='binary_r')
```
