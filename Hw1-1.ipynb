```python
%matplotlib inline
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def blend(a, b, alpha):
    '''Blends two images using the blend factor
    Args: 
        a (numpy.array) - image A
        b (numpy.array) - image B
        alpha (float) - weight factor
    Returns: (numpy.array) - weighted image'''
    return cv2.add(alpha*a, (1-alpha)*b)
```

```python
dolphin = cv2.imread('images/dolphin.png', cv2.IMREAD_GRAYSCALE)
bicycle = cv2.imread('images/bicycle.png', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(dolphin, cmap='binary_r')
```

```python
plt.imshow(bicycle, cmap='binary_r')
```

```python
blended = blend(dolphin, bicycle, 0.75)
plt.imshow(blended, cmap='binary_r')
```
